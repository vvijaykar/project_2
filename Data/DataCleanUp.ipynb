```python
# Dependencies & set up
import pandas as pd
import numpy as np

data_csv = "Data/csv/mainbehaviordata.csv"
```

```python
main_df = pd.read_csv(data_csv)
```

## Checkout dataframe

```python
print(len(main_df))
print(main_df.dtypes)
main_df.head()
```

## Find shops with and without FES

```python
shops_no_fes = main_df.loc[(main_df['Behavior'] == 'FES Present or Found') & (main_df['Received'] == 0)]
```

```python
shops_no_fes
```

```python
shops_w_fes = main_df.loc[(main_df['Behavior'] == 'FES Present or Found') & (main_df['Received'] > 0)]
```

```python
shops_w_fes
```

```python
shops_no_fes = shops_no_fes['shop_num']
```

```python
shops_w_fes = shops_w_fes['shop_num']
```

## Separate the main df to see data for shops w/ FES and w/o FES

```python
no_fes_data = pd.merge(shops_no_fes, main_df, on='shop_num', how='left')
```

```python
no_fes_data.head()
```

```python
no_fes_data['FES'] = 'No FES Present'
```

```python
no_fes_data.head()
```

```python
w_fes_data = pd.merge(shops_w_fes, main_df, on='shop_num', how='left')
```

```python
w_fes_data['FES'] = 'FES Present'
```

```python
w_fes_data.head()
```

## Put them back together

```python
dfs = [no_fes_data, w_fes_data]
```

```python
concat_df = pd.concat(dfs)
```

```python
print(len(concat_df))
concat_df.head()
```

## Group by main fields and run aggs

```python
grouped_df = concat_df.groupby(['FES', 'ST', 'MSA', 'City', 'StoreID', 'Behavior'], as_index=False).agg({'Received': 'sum', 'Possible': 'sum'})
```

```python
grouped_df['Percent'] = grouped_df['Received']/grouped_df['Possible']
```

```python
grouped_df
```

```python
grouped_df.to_csv('Data/csv/final_fes_data.csv')
```
